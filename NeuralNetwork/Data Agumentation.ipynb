```python
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from numpy import expand_dims
from matplotlib import pyplot as plt
```

```python
help(ImageDataGenerator)
```

```python
img = load_img('bird.jpg')
_img = img_to_array(img)
samples = expand_dims(_img, 0)
```

### Horizontal and Vertical Shift Augmentation

#### Horizontal Shift 

```python
data_gen = ImageDataGenerator(width_shift_range=[-200, 200])
it = data_gen.flow(samples, batch_size=1)

for i in range(9):
    plt.subplot(330+1+i)
    batch = it.next()
    image = batch[0].astype('uint8')
    plt.imshow(image)
```

#### Vertical shift: 

```python
data_gen = ImageDataGenerator(height_shift_range=[0, 0.6])
it = data_gen.flow(samples, batch_size=1)

for i in range(9):
    plt.subplot(330+1+i)
    batch = it.next()
    _img = batch[0].astype('uint8')
    plt.imshow(_img)
```

```python
data_gen = ImageDataGenerator(height_shift_range=[0, 0.2])
it = data_gen.flow(samples, batch_size=1)

for i in range(9):
    plt.subplot(330+1+i)
    batch = it.next()
    _img = batch[0].astype('uint8')
    plt.imshow(_img)
```

```python
data_gen = ImageDataGenerator(height_shift_range=[0, 0.8])
it = data_gen.flow(samples, batch_size=1)

for i in range(9):
    plt.subplot(330+1+i)
    batch = it.next()
    _img = batch[0].astype('uint8')
    plt.imshow(_img)
```

### Horizontal and Vertical Flip Augmentation

#### Horizontal Flip: 

```python
data_gen = ImageDataGenerator(horizontal_flip=True)
it = data_gen.flow(samples, batch_size=1)

for i in range(9): 
    plt.subplot(330+1+i)
    batch = it.next()
    img = batch[0].astype('uint8')
    plt.imshow(img)
```

#### Vertical Flip: 

```python
data_gen = ImageDataGenerator(vertical_flip=True)
it = data_gen.flow(samples, batch_size=1)

for i in range(9):
    plt.subplot(330+1+i)
    batch = it.next()
    plt.imshow(batch[0].astype('uint8'))
```

### Random Rotation Augmentation

```python
data_gen = ImageDataGenerator(rotation_range=180)
it = data_gen.flow(samples, batch_size=1)

for i in range(9):
    plt.subplot(330+1+i)
    batch = it.next()
    plt.imshow(batch[0].astype('uint8'))
```

### Random Brightness Augmentation

Any value below 1 will darken the image and and above 1 will brighten the image. It takes 1 as a base line of the image. This will be helpful to identify the image in low light and brighter light also.

```python
data_gen = ImageDataGenerator(brightness_range=[2, 0.4])
it = data_gen.flow(samples, batch_size=1)

for i in range(9):
    plt.subplot(330+1+i)
    batch = it.next()
    plt.imshow(batch[0].astype('uint8'))
```

### Random Zoom Augmentation

A zoom augmentation randomly zooms the image in and either adds new pixel values around the image or interpolates pixel values respectively.

If a float is specified, then the range for the zoom will be [1-value, 1+value]. For example, if you specify 0.3, then the range will be [0.7, 1.3], or between 70% (zoom in) and 130% (zoom out).

```python
data_gen = ImageDataGenerator(zoom_range=[0.6, 1.4])
it = data_gen.flow(samples, batch_size=1)

for i in range(9):
    plt.subplot(330+1+i)
    batch = it.next()
    plt.imshow(batch[0].astype('uint8'))
```
