```python
# Get the percentage of missing values in each column
round ( (100 * ( housing_data.isnull().sum() / len(housing_data.index) ) ), 2 ) 
```

```python
housing_data = housing_data.drop('BuildingArea', axis=1)
housing_data = housing_data.drop('YearBuilt', axis=1)
housing_data = housing_data.drop('CouncilArea', axis=1)
housing_data.head()
```

```python
housing_data[housing_data.isnull().sum(axis=1) > 5] 
```

```python
len(housing_data[housing_data.isnull().sum(axis=1) > 5].index)
```

```python
housing_data = housing_data[housing_data.isnull().sum(axis=1) < 5]
round ( (100 * ( housing_data.isnull().sum() / len(housing_data.index) ) ), 2 ) 
```

```python
## Remove the rows which price has been null
housing_data = housing_data[-np.isnan(housing_data['Price'])]
round ( (100 * ( housing_data.isnull().sum() / len(housing_data.index) ) ), 2 ) 
```

```python
housing_data['Landsize'].describe()
```

```python
housing_data= housing_data[-np.isnan(housing_data['Landsize'])]
round ( (100 * ( housing_data.isnull().sum() / len(housing_data.index) ) ), 2 ) 
```

```python
housing_data.loc[:, ['Lattitude', 'Longtitude']].describe()
```

```python
housing_data.loc[np.isnan(housing_data['Lattitude']), ['Lattitude']] = housing_data['Lattitude'].mean
housing_data.loc[np.isnan(housing_data['Longtitude']), ['Longtitude']] = housing_data['Longtitude'].mean
```

```python
round ( (100 * ( housing_data.isnull().sum() / len(housing_data.index) ) ), 2 ) 
```

```python
housing_data.loc[:, ['Bathroom', 'Car']].describe()
```

```python
housing_data['Bathroom'].mode()
```

```python
housing_data['Car'] = housing_data['Car'].astype('category')
```

```python
housing_data['Car'].value_counts()
```

```python
housing_data.loc[pd.isnull(housing_data['Car']), ['Car']] = 2
```

```python
round ( (100 * ( housing_data.isnull().sum() / len(housing_data.index) ) ), 2 ) 
```

```python
housing_data['Bathroom'] = housing_data['Bathroom'].astype('category')
housing_data['Bathroom'].describe()
```

```python
housing_data['Bathroom'].value_counts()
```

```python
housing_data.loc[pd.isnull(housing_data['Bathroom']), ['Bathroom']] = 1
```

```python
round ( (100 * ( housing_data.isnull().sum() / len(housing_data.index) ) ), 2 ) 
```
